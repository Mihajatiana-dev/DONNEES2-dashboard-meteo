```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading raw data
df = pd.read_csv("../dags/data/historical/historical.csv", parse_dates=["datetime"])
print(f"Data period: {df['datetime'].min()} to {df['datetime'].max()}")
print(f"Dimensions : {df.shape}\nCities : {df['name'].unique().tolist()}")
```

```python
# Columns
print("Dataset columns")
print(df.columns.tolist())
```

```python
# Data overview
df.head()
```

```python
# Data distribution by city
df["name"].value_counts().plot(kind="bar", color="green")
plt.title("Number of observations per city");
plt.ylabel("number of data")
```

```python
# Missing values
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing values by columns");
```

```python
# Temperature by city (average temperature)
plt.figure(figsize=(14, 8))
for ville in df["name"].unique():
    df_ville = df[df["name"] == ville].sort_values("datetime")
    plt.plot(df_ville["datetime"], df_ville["temp"], label=ville, alpha=0.7)
plt.xlabel("date")
plt.ylabel("temperature (Â°C)")
plt.title("Temperature trends by city")
plt.legend()
plt.grid(True);
```

```python
# All weather conditions
all_conditions = df["conditions"].value_counts()

fig, ax = plt.subplots(figsize=(15, 10))
wedges, texts = ax.pie(all_conditions.values, startangle=90)

percentages = (all_conditions.values / all_conditions.sum()) * 100
legend_labels = [f"{condition} - {percent:.1f}%" for condition, percent in zip(all_conditions.index, percentages)]

ax.legend(wedges, legend_labels, title="Weather conditions", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title("Distribution of all weather conditions")
plt.tight_layout()
```
